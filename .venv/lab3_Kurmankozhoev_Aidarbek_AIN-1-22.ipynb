```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем произвольную квадратичную функцию
def quadratic_function(x):
    return 2 * x**2 - 3 * x

# Генерируем значения для x
x = np.linspace(-10, 10, 400)

# Генерируем значения для y по функции
y = quadratic_function(x)

# Создаем график
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = 2x^2 - 3x + 5", color="blue")

# Настройки подписей осей и заголовка
plt.title("График квадратичной функции", fontsize=16)
plt.xlabel("Значения X", fontsize=12)
plt.ylabel("Значения Y", fontsize=12)

# Добавляем сетку
plt.grid(True)

# Добавляем легенду
plt.legend()

# Отображаем график
plt.show()
```

```python
# Определяем произвольную гармоническую функцию
def harmonic_function(x):
    return np.tan(2 * np.pi * x)

# Генерируем значения для x
x = np.linspace(0, 10, 400)

# Генерируем значения для y по функции
y = harmonic_function(x)

# Создаем график
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = tan(2πx)", color="green")

# Настройки подписей осей и заголовка
plt.title("График гармонической функции", fontsize=16)
plt.xlabel("Значения X", fontsize=12)
plt.ylabel("Значения Y", fontsize=12)

# Добавляем сетку
plt.grid(True)

# Добавляем легенду
plt.legend()

# Отображаем график
plt.show()
```

```python
# Создаем графики с использованием функции subplot()

# Создаем фигуру с двумя подграфиками (2 строки, 1 колонка)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Первый график: квадратичная функция
ax1.plot(x, quadratic_function(x), label="y = 2x^2 - 3x", color="blue")
ax1.set_title("График квадратичной функции", fontsize=16)
ax1.set_xlabel("Значения X", fontsize=12)
ax1.set_ylabel("Значения Y", fontsize=12)
ax1.grid(True)
ax1.legend()

# Второй график: гармоническая функция
ax2.plot(x, harmonic_function(x), label="y = tan(2πx)", color="green")
ax2.set_title("График гармонической функции", fontsize=16)
ax2.set_xlabel("Значения X", fontsize=12)
ax2.set_ylabel("Значения Y", fontsize=12)
ax2.grid(True)
ax2.legend()

# Настройка отображения графиков
plt.tight_layout()
plt.show()
```

```python
# Создаем график на одной координатной плоскости
plt.figure(figsize=(8, 6))

# График квадратичной функции
plt.plot(x, quadratic_function(x), label="y = 2x^2 - 3x", color="blue")

# График гармонической функции
plt.plot(x, harmonic_function(x), label="y = tan(2πx)", color="green")

# Настройки подписей осей и заголовка
plt.title("Графики квадратичной и гармонической функций", fontsize=16)
plt.xlabel("Значения X", fontsize=12)
plt.ylabel("Значения Y", fontsize=12)

# Добавляем сетку
plt.grid(True)

# Добавляем легенду
plt.legend()

# Отображаем график
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
t = np.linspace(0, 4 * np.pi, 500)

# Параметрические функции
x = np.cos(3 * t)
y = np.sin(2 * t)

# Построение графика
plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.title("Параметрический график: x(t) = cos(3t), y(t) = sin(2t)")
plt.xlabel("x(t) = cos(3t)")
plt.ylabel("y(t) = sin(2t)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(values, labels=years, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.title('Круговая диаграмма прожиточного минимума в КР (2012-2022)')
plt.axis('equal')  # Равные оси для кругового графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание сетки координат
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.cos(x) * np.sin(y)  # Другая произвольная функция

# Построение каркасной поверхности
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z, color='black')

# Настройка осей и заголовков
ax.set_title('Каркасная поверхность: z = cos(x) * sin(y)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
